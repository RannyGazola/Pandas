```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv')
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
df
```

```python
anos = list(map(str, range(1980,2014)))
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil_dic = {'ano':brasil.index.tolist(), 'imigrantes' : brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dic)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

## Desafio

```python
argentina = df.loc['Argentina', anos]
argentina
```

```python
argentina_dict = {'ano':argentina.index.tolist(), 'imigrantes' : argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
dados_argentina
```

```python
plt.figure(figsize=(8,4))
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'], label='Brasil', color='Purple')
plt.plot(dados_argentina['ano'],dados_argentina['imigrantes'], label='Argentina',color='Green')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```
