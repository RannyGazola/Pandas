```python
import pandas as pd
import numpy as np

dados = pd.read_json('dados_vendas_clientes.json')
dados = pd.json_normalize(dados['dados_vendas'])
colunas = list(dados.columns)
dados = dados.explode(colunas[1:])
dados.reset_index(inplace=True, drop=True)
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','').strip())
dados['Valor da compra'] = dados['Valor da compra'].astype('int64')
dados['Cliente'] = dados['Cliente'].str.upper().replace('[^a-zA-Z\-\'\Ã\ ]','',regex=True)
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'])
dados.sort_values('Valor da compra', ascending=False)
```
